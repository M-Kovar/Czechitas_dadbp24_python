# Cvičení 1: Vláknina

## Zadání

Problémem dnešních potravin je často nedostatek vlákniny. Konzumace potravin bohatých na vlákninu může být pro řadu lidí zdravotně prospěná. Vyhledej v tabulce `food_nutrient` potraviny, které obsahují alespoň 10 gramů vlákniny. Vláknina je uložená pod názvem `Fiber, total dietary`. Napiš dotaz jako jeden příkaz a využij operátor `&`.

```python
import pandas as pd

food_nutrient = pd.read_csv("food_nutrient.csv")
```

```python
fiber_rich = food_nutrient[
    (food_nutrient["name"] == "Fiber, total dietary")
    & (food_nutrient["amount"] >= 10)
]

# Pomoci metody query
# fiber_rich = food_nutrient.query('name == "Fiber, total dietary" and amount >= 10')

fiber_rich
```

# Cvičení 3: Česká jména

Stáhni a načti si soubor [jmena.csv](https://kodim.cz/cms/assets/czechitas/python-data-1/python-pro-data-1/podmineny-vyber/podmineny-vyber/ceska-jmena/jmena.csv), která obsahuje nejpoužívanější česká jména.

- Vypiš všechny řádky se jmény, jejichž nositelé mají průměrný věk vyšší než 60 (hodnota ve sloupci `prumerny_vek` je větší než 60).
- Vypiš pouze jména z těch řádků, kde četnost je mezi 80 000 a 100 000.
- Vypiš jména a četnost pro jména se slovanským nebo hebrejským původem. Kolik takových jmen je?

Pro poslední úkol můžeš využít operátor `|`. Alternativně si můžeš vyzkoušet metodu `.isin()`, která zápis zkrátí. 
Jako parametr vkládáme seznam hodnot, které vyhovují podmínce. Níže je příklad použití metody.  
Z tabulky `tabulka` chceme vybrat řádky, které ve sloupci `sloupec` mají hodnotu `hodnota_1` nebo `hodnota_2`. 
Pokud jsi vytvořil(a) i verzi s operátorem `|`, můžeš obě verze porovnat a rozhodnout se, která verze se ti líbí více.

`tabulka = tabulka[tabulka["sloupec"].isin(["hodnota_1", "hodnota_2"])]`

```python
import pandas as pd

# Stažení z URL
url = "https://kodim.cz/cms/assets/czechitas/python-data-1/python-pro-data-1/podmineny-vyber/podmineny-vyber/ceska-jmena/jmena.csv"
jmena = pd.read_csv(url)

# Alternativa: načtení předem uloženého souboru
# jmena = pd.read_csv("jmena.csv")

jmena
```

```python
# Vypiš všechny řádky se jmény, jejichž nositelé mají průměrný věk vyšší než 60.
jmena[jmena["prumerny_vek"] > 60]
```

```python
# Vypiš pouze jména z těch řádků, kde četnost je mezi 80 000 a 100 000 - pomocí složené podmínky
jmena[(jmena["cetnost"] >= 80_000) & (jmena["cetnost"] <= 100_000)]
```

```python
# Vypiš pouze jména z těch řádků, kde četnost je mezi 80 000 a 100 000 - pomocí složené podmínky
jmena[jmena["cetnost"].between(80_000, 100_000)]
```

```python
# Slovanský nebo Hebrejský původ - pomocí složené podmínky
jmena[(jmena["puvod"] == "slovanský") | (jmena["puvod"] == "hebrejský")]
```

```python
# Slovanský nebo Hebrejský původ - pomocí isin()
jmena[jmena["puvod"].isin(["slovanský", "hebrejský"])]
```
