```python
import pandas as pd

titanic = pd.read_csv('titanic.csv')
```

```python
# Nachystame data pro vizualizaci
survivor_counts = titanic['Survived'].value_counts()
survivor_counts
```

```python
# Pocty prezivsich cestujicich
survivor_counts.plot(kind="bar")

# survivor_counts.plot(kind="bar", title="Poƒçet cestuj√≠c√≠ch podle p≈ôe≈æit√≠", xlabel="P≈ôe≈æil", ylabel="Poƒçet", grid=True)
```
