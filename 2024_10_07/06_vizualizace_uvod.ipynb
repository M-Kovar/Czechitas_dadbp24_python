# Vizualizace - ochutnávka

```python
import pandas as pd

titanic = pd.read_csv('titanic.csv')
```

```python
# Nachystame data pro vizualizaci
survivor_counts = titanic['Survived'].value_counts()
survivor_counts
```

```python
# Pocty prezivsich cestujicich
survivor_counts.plot(kind="bar")

# survivor_counts.plot(kind="bar", title="Počet cestujících podle přežití", xlabel="Přežil", ylabel="Počet", grid=True)
```

```python
# Rozdeleni cestujicich podle pohlavi, s vyuzitim retezeni metod
titanic['Sex'].value_counts().plot(kind="pie")
```
