# Podmíněný výběr

```python
import pandas as pd

food_nutrient = pd.read_csv("food_nutrient.csv")
food_nutrient
```

```python
# Pouze pro nazornost: zobrazeni vystupu podminky
food_nutrient["name"] == "Magnesium, Mg"
```

```python
# Vytvoreni noveho DF, ktery obsahuje pouze radky o horciku

# Analogie SQL: SELECT * FROM food_nutrient WHERE name = 'Magnesium, Mg'
magnesium = food_nutrient[food_nutrient["name"] == "Magnesium, Mg"]

magnesium
# magnesium.shape
```

```python
magnesium["amount"].describe()
```

```python
# Analogie SQL: SELECT * FROM magnesium WHERE amount > 100
magnesium_limit = magnesium[magnesium["amount"] > 100]
magnesium_limit
```

```python
calcium = food_nutrient[food_nutrient["name"] == "Calcium, Ca"]
calcium
```

```python
calcium_limit = calcium[(calcium["amount"] > 30) & (calcium["amount"] < 500)]

# Prohledneme si zaznamy se tremi nejvyssimi hodnotami
calcium_limit.nlargest(3, "amount")
```

# Bonus: Výběr metodou query()

Alternativa rovnocenná zápisu podmíněného výběru výše. Odlišná syntaxe více se podobající SQL. Při výběru pomocí většího množství podmínek zpřehledňuje zápis.

```python
calcium_limit = calcium.query('amount > 30 and amount < 500')

# Jeste jednodussi zapis (preferovany)
# calcium_limit = calcium.query('30 < amount < 500')

# calcium_limit.shape
```

```python
# Bonus z bonusu: Spojeni vyberu nutrientu a jeho mnozstvi do jednoho dotazu
calcium_limit = food_nutrient.query('name == "Calcium, Ca" and 30 < amount < 500')
# calcium_limit.shape
```
