# Pokročilé úpravy

## Chybějící hodnoty

```python
import pandas as pd
food_nutrient = pd.read_csv('food_nutrient.csv')

food_nutrient
```

```python
food_nutrient.info()
```

```python
food_nutrient.isna().sum()
```

```python
food_nutrient_amount_isNaN = food_nutrient[food_nutrient["amount"].isna()]
food_nutrient_amount_isNaN
```

```python
food_nutrient_amount_isFilled = food_nutrient[food_nutrient["amount"].notna()]
food_nutrient_amount_isFilled
```

```python
# Odstraní řádky, ve kterých se nachází aspoň jeden NaN  (tzn. aspoň jedna hodnota chybí)
food_nutrient.dropna()
```

```python
# Odstraní řádky, které obsahují NaN ve sloupci "amount"
food_nutrient = food_nutrient.dropna(subset="amount")
food_nutrient.shape
```

## Spojení dat

```python
food_sample_100 = pd.read_csv("food_sample_100.csv")

display(food_sample_100.head())
food_sample_100.shape
```

```python
food_other = pd.read_csv("food_other.csv")

display(food_other.head())
food_other.shape
```

```python
# Záznamy spojené pod sebe si zachovají původní indexy
food = pd.concat([food_sample_100, food_other])
food.to_csv("food.csv", encoding="utf-8")
```

```python
# ignore_index=True ignoruje původní indexy a vytvoří nový
food = pd.concat([food_sample_100, food_other], ignore_index=True)
food
```

## Propojení dat

```python
display(food.head())
food.shape
```

```python
display(food_nutrient.head())
food_nutrient.shape
```

```python
food_merged = pd.merge(food, food_nutrient, on="fdc_id")

display(food_merged.head())
food_merged.shape
```
