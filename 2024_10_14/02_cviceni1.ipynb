# Cvičení 1: Pokročilé úpravy - spojování

```python
import pandas as pd

# Příprava vstupního DataFramu
food_nutrient = pd.read_csv("food_nutrient.csv")
food_sample_100 = pd.read_csv("food_sample_100.csv")
food_other = pd.read_csv("food_other.csv")
food = pd.concat([food_sample_100, food_other], ignore_index=True)
food_merged = pd.merge(food, food_nutrient, on="fdc_id")
```

V tabulce je sloupec `fdc_id`, pomocí kterého ji můžeš propojit s tabulkou `food_merged`. Protože názvy jsou v obou
tabulkách stejné, takže by bylo možné použít parametr `on`. Vyzkoušej si ale místo toho parametry `left_on`
a `right_on`, kterým dáš stejnou hodnotu, tj. název sloupce `fdc_id`. Výsledek ulož do tabulky `food_merged_brands`.

```python
food_merged_brands = pd.merge(food_merged, branded_food, left_on="fdc_id", right_on="fdc_id")
# ... pokud nevyplníme parametr "how", použije se výchozí hodnota inner

food_merged_brands.head()
```

Pro tabulku `food_merged_brands` splň následující úkoly.

- U výsledné tabulky `food_merged_brands` zkontroluj počet řádků a srovnej ho s původní tabulkou `food_merged`. Ubyly
  nějaké řádky? A čím to je?
- Nyní proveď operaci `merge` znovu, ale s parametrem `how` nastaveným na hodnotu `left`. Zkontroluj počet řádků a
  porovnej ho s počtem řádků tabulky `food_merged_brands`. Proč se počet liší?

```python
print(f"Shape branded_food: {branded_food.shape}")
print(f"Shape food_merged: {food_merged.shape}")
```

```python
food_merged_brands_left = pd.merge(food_merged, branded_food, left_on="fdc_id", right_on="fdc_id", how="left")
food_merged_brands_left
```

```python
print(f"Shape branded_food: {branded_food.shape}")
print(f"Shape food_merged: {food_merged.shape}")
print(f"Shape food_merged_brands (inner merge): {food_merged_brands.shape}")
print(f"Shape food_merged_brands_left (left merge): {food_merged_brands_left.shape}")
```

**...Proč se počet liší?**
- `food_merged_brands` spojeno výchozí metodou `inner`
    - -> výsledek obsahuje pouze záznamy (řádky), pro které se `fdc_id` shoduje v obou vstupních tabulkách (průnik). Tím pádem výsledek obsahuje jen záznamy, které mají jak informace o potravině (DF food_merged), tak informace o značce (DF branded_food).
    
- `food_merged_brands_left` spojeno metodou `left`
    - -> výsledek obsahuje záznamy (řádky), pro které se `fdc_id` shoduje v obou vstupních tabulkách plus všechny ostatní záznamy (řádky) z levé (první) vstupní tabulky; záznamy z levé tabulky, pro které nebyl nalezen záznam v pravé tabulce, budou zprava doplněny sloupci s prázdnými hodnotami (NaN). Konkrétně jde o potraviny (DF food_merged), ke kterým nebyly nalezeny informace o značce (DF branded_food).

- ... Nicméně zrovna v tomto příkladu se počty neliší, jelikož při využití obou metod jsou úspěšně spárovány všechny záznamy z levé tabulky.
